```python
people={ 
    'first':['corey','jane','john','chirs',np.nan,None,'NA'],
    'last':['Schafer','Doe','doe','Schafer',np.nan,np.nan,'Missing'],
    'email': ['coreyschafe@gmail.com','janedoe@gmail.com','johndoe@gmail.com',None,np.nan,'Missing','NA'],
    'age':['33','45','65',None,None,'Missing','NA']}
```
